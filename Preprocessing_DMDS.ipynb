```python
import numpy as np
import pandas as pd
import warnings
```

```python
df_netflix=pd.read_csv("netflix_titles.csv")
df_disney=pd.read_csv("disney_plus_titles.csv")
```

# Add Column [Platform]

```python
# Tambah kolom platform di tiap dataset
df_netflix['Platform'] = "Netflix"
df_disney['Platform'] = "Disney+"
df_netflix
```

# Merge Netflix and Disney+ Datasets into 1 Dataset

```python
# Merge 2 dataset menjadi 1 dan drop kolom yang tidak dibutuhkan
df = df_netflix.append(df_disney, ignore_index=True)
df = df.rename(columns={'show_id': 'Show ID',
                        'type': 'Show Type', 
                        'director': 'Director', 
                        'cast': 'Cast',
                        'type': 'Show Type',
                        'country':'Production Country', 
                        'date_added':'Date Added', 
                        'release_year':'Release Year', 
                        'rating':'Age Rating',
                        'duration': 'Duration',
                        'listed_in': 'Genres',
                        'description': 'Description'})

df
```

```python
df.count()
```

# Remove Rows with NaN Value on Column

```python
# Drop rows dengan nilai "date_added", "Age Rating", & "Duration" yang NaN
df.dropna(subset = ["Date Added"], inplace=True)
df.dropna(subset = ["Age Rating"], inplace=True)
df.dropna(subset = ["Duration"], inplace=True)
df.reset_index(drop=True, inplace=True)

df.count()
```


# Fill Column [Production Country] NaN Value with "No Data"

```python
# Ubah semua value NaN menjadi "No Data", agar bisa di pre-processing
df['Director'].replace(np.nan, "No Data", inplace=True)
df['Cast'].replace(np.nan, "No Data", inplace=True)
df['Production Country'].replace(np.nan, "No Data", inplace=True)

df.count()
```

# Transform Column [Production Country]

```python
# Ambil value kolom "country" diubah ke list
country_list = df['Production Country'].to_list()
len(country_list)

# Ubah setiap value country_list menjadi 1 negara saja (Negara Paling Pertama)
for x in range(len(country_list)):
  country_list[x] = country_list[x].split(',')[0]
country_list

# Replace kolom "country" dengan "country_list" yang sudah dibuat
df['Production Country'] = country_list

df
```

# Convert Column [Date Added] to Date

```python
# Convert "date_added" variable type to datetime
import datetime
df['Date Added'] = pd.to_datetime(df['Date Added'], dayfirst=True)

df
```

# Convert Column [Release Year] to Date

```python
# Ambil value kolom "release_year" diubah ke list
date_list = df['Release Year'].to_list()
len(date_list)

# Ubah semua value "release_year" menjadi 30 Juni + "release_year"
for x in range(len(date_list)):
  date_list[x] = 'June 30, ' + str(date_list[x]).split('-')[0]
date_list

# Replace kolom "release_year" dengan "date_list" yang telah dibuat
df['Release Year'] = date_list

# Convert "release_year" variable type to datetime
df['Release Year'] = pd.to_datetime(df['Release Year'], dayfirst=True)

df
```

# Add Column [Up to Date]

```python
# Membuat list up_to_date
up_to_date = []

# Mengisi value dengan "Up to Date" jika "date_added"-"release_year" < 365
for x in range(len(df['Date Added'])):
  sum_date = df['Date Added'][x]-df['Release Year'][x]
  sum_date = str(sum_date).split(' ')[0]
  if int(sum_date) <= 365:
    up_to_date.append(True)
  else:
    up_to_date.append(False)

# Menambahkan kolom "Up to Date" dengan value sesuai dengan list up_to_date
df['Up to Date (Added Less than a Year from Release Year)'] = up_to_date
df
```

# Data Discretization on Column [Age Rating]

```python
warnings.filterwarnings("ignore")
for x in range(len(df['Age Rating'])):
    if df['Age Rating'][x] == 'G' or df['Age Rating'][x] == 'TV-G' or df['Age Rating'][x] == 'UR' or df['Age Rating'][x] == 'NR' or df['Age Rating'][x] == '66 min' or df['Age Rating'][x] == '74 min' or df['Age Rating'][x] == '84 min':
        df['Age Rating'][x] = 'All Age'
    elif df['Age Rating'][x] == 'TV-Y' or df['Age Rating'][x] == 'TV-Y7' or df['Age Rating'][x] == 'TV-Y7-FV':
        df['Age Rating'][x] = 'Kids (<12 Years Old)'
    elif df['Age Rating'][x] == 'PG' or df['Age Rating'][x] == 'PG-13' or df['Age Rating'][x] == 'TV-PG' or df['Age Rating'][x] == 'TV-14' or df['Age Rating'][x] == 'NC-17':
        df['Age Rating'][x] = 'Teenagers (13-18 Years Old)'
    elif df['Age Rating'][x] == 'NC' or df['Age Rating'][x] == 'R' or df['Age Rating'][x] == 'TV-MA':
        df['Age Rating'][x] = 'Adults (>18 Years Old)'
df
```

# Splitting Column [Genres] from Multivalue to Columns

```python
# Splitting Multivalue column (genre / listed_in)
df_genre_ = df['Genres'].str.split(",", expand=True)

df_genre_
```

# Data Discretization Column [Genres]

```python
for column in range(df_genre_.shape[1]):
    for i in range(df_genre_.shape[0]):
        x = df_genre_.loc[i][column]
        if x is not None: x = x.strip()
            
        if x == 'Comedies' or x == 'TV Comedies' or x == 'Stand-Up Comedy & Talk Shows' or x == 'Stand-Up Comedy' or x == 'Parody':
            df_genre_.loc[i][column] = 'Comedy'
        elif x == 'Anime Features' or x == 'Anime Series':
            df_genre_.loc[i][column] = 'Anime'
        elif x == 'Romantic Movies' or x == 'Romantic TV Shows' or x == 'Romantic Comedy':
            df_genre_.loc[i][column] = 'Romance'
        elif x == 'Documentaries' or x == 'Docuseries' or x == 'Biographical':
            df_genre_.loc[i][column] = 'Documentary'
        elif x == 'Sports Movies':
            df_genre_.loc[i][column] = 'Sports'
        elif x == "Kids' TV":
            df_genre_.loc[i][column] = 'Kids'
        elif x == 'Children & Family Movies' or x == 'Buddy':
            df_genre_.loc[i][column] = 'Family'
        elif x == 'TV Action & Adventure' or x == 'Action & Adventure':
            df_genre_.loc[i][column] = 'Action-Adventure'
        elif x == 'Dramas' or x == 'TV Dramas' or x == 'Soap Opera / Melodrama':
            df_genre_.loc[i][column] = 'Drama'
        elif x == 'Crime TV Shows' or x == 'Crime' or x == 'Spy/Espionage' or x == 'Police/Cop':
            df_genre_.loc[i][column] = 'Crime'
        elif x == 'TV Thrillers' or x == 'Thrillers' or x == 'TV Horror' or x == 'Horror Movies' or x == 'Thriller':
            df_genre_.loc[i][column] = 'Thriller & Horror'
        elif x == 'Music & Musicals' or x == 'Musical' or x == 'Concert Film':
            df_genre_.loc[i][column] = 'Music'
        elif x == 'Series' or x == 'TV Shows' or x == 'International TV Shows' or x == 'Korean TV Shows' or x == 'Spanish-Language TV Shows' or x == 'British TV Shows' or x == 'Western' or x == 'Teen TV Shows':
            df_genre_.loc[i][column] = None
        elif x == 'Independent Movies' or x == 'International Movies' or x == 'Classic Movies' or x == 'Movies':
            df_genre_.loc[i][column] = None
        elif x == 'Lifestyle' or x == 'Reality' or x == 'Travel' or x == 'Faith & Spirituality':
            df_genre_.loc[i][column] = 'Lifestyle'
        elif x == 'Dance' or x == 'Game Show / Competition' or x == 'Talk Show' or x == 'Anthology' or x == 'Reality TV' or x == 'Variety':
            df_genre_.loc[i][column] = 'Entertainment'
        elif x == 'Cult Movies' or x == 'Classic & Cult TV':
            df_genre_.loc[i][column] = 'Cult'
        elif x == 'Mystery' or x == 'TV Mysteries':
            df_genre_.loc[i][column] = 'Mystery'
        elif x == 'Disaster' or x == 'Animals & Nature' or x == 'Survival' or x == 'Science & Nature TV':
            df_genre_.loc[i][column] = 'Nature'
        elif x == 'Fantasy' or x == 'Coming of Age' or x == 'Superhero' or  x == 'Sci-Fi & Fantasy' or x == 'TV Sci-Fi & Fantasy':
            df_genre_.loc[i][column] = 'Science Fiction'

df_genre_
```

# Merge the Columns Back and Replace Column [Genres] with New Genres

```python
new_genre_list = []
for row in range(df_genre_.shape[0]):
    temp = ''
    for x in range(df_genre_.shape[1]):
        if df_genre_.loc[row][x] is not None:
            if temp != '':
                temp += ', '
            temp += df_genre_.loc[row][x]

    new_genre_list.append(temp.strip())
    
df['Genres'] = new_genre_list
df
```

```python
df=pd.read_csv("Dataset.csv")
```

```python
df.rename(columns = {'Show ID':'ShowID', 'Show Type': 'ShowType', 'title': 'Title', 'Production Country': 'ProductionCountry', 'Date Added': 'DateAdded', 'Release Year': 'ReleaseYear', 'Age Rating': 'AgeRating'}, inplace = True)
```

```python
# re-index
df = df.reset_index()
df.index = df.index + 1
df['ShowID'] = 's'+df['index'].astype(str)
```

```python
df = df[['ShowID', 'ShowType', 'Title', 'Director', 'Cast', 'ProductionCountry', 'DateAdded',
                   'ReleaseYear', 'AgeRating', 'Duration', 'Genres', 'Description', 'Platform']]
df
```

```python
# buat table country & rating
country = pd.DataFrame()
country['ProductionCountry'] = df.ProductionCountry.unique()
rating = pd.DataFrame()
rating['AgeRating'] = df.AgeRating.unique()
```

```python
from sklearn import preprocessing
pc = preprocessing.LabelEncoder()
pc.fit(df['ProductionCountry'])
country['CountryID'] = pc.fit_transform(country['ProductionCountry'])
df['ProductionCountryID'] = pc.fit_transform(df['ProductionCountry'])
country['CountryID'] = country['CountryID'] + 1
country = country.sort_values('CountryID')
country = country[['CountryID', 'ProductionCountry']]
df['ProductionCountryID'] = df['ProductionCountryID'] + 1
```

```python
ar = preprocessing.LabelEncoder()
ar.fit(df['AgeRating'])
rating['RatingID'] = ar.fit_transform(rating['AgeRating'])
df['AgeRatingID'] = ar.fit_transform(df['AgeRating'])
df['AgeRatingID'] = df['AgeRatingID'] + 1
rating['RatingID'] =rating['RatingID']  + 1
rating = rating.sort_values('RatingID')
rating = rating[['RatingID', 'AgeRating']]
```

```python
movies = df[['ShowID', 'ShowType', 'Title', 'DateAdded', 'ReleaseYear', 'Duration', 'Description', 'Platform', 'ProductionCountryID', 'AgeRatingID']]
```

```python
# mysql
movies.to_csv('movies.csv', index=False, header=False)
country.to_csv('country.csv', index=False, header=False)
rating.to_csv('rating.csv', index=False, header=False)
```
